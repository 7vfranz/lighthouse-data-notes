Preferential Attachment
* property of some networks
* as they grow and evolve - majority of new edges are to nodes that are already in high-degree (already connected to a lot)
* degree of these nodes increases faster than other nodes
* resulting in small nodes with high degree - and large amount of nodes with low degree (long-tailed degree distribution)
* based on:
    * popularity - 'rich gets richer'
    * quality - evaluate people and everything based on objective quality 'good gets better'
    * mixed  - among nodes of similar attributes - one become 'stars' with many friends/followers - halo effect

Graph Theory Algorithms
* Centrality
    * Degree, Betweeness, Closeness
    * calcualte the importance of a particular ndoe 
    * combination of these three - high/low can make for different interactions
    * see link
    * Degree
        * count of how many edges are connected to each node
        * higher degree = node with more edges 
        * in-degree - alot of nodes going towards the node
        * out-degree - reaches out to many other nodes
        * essentially id nodes with most connections to other nodes 
    * Betweenness Centrality
        * how often a particular node lies on a shortest path between two other nodes
        * a 'connector' - removing that node would increase distance between a lot of other nodes
        * e.g. layover airports
        * contains the most information about a network
    * Closeness Centrality
        * proximity of nodes to other nodes
        * high closeness - spreads quickly
        * airports with highcloseness - lots of quick small flights
        * average length of shortest path from node to other nodes
    * Page Rank
        * used by google to rank websites
        * neuron - pageRank correlate with relative firing rate 
        * good for citations and authority
    * Eigencentrality
        * measures node influence on number of links to other nodes
        * how well connected a node is and how many links their connections have
        * identify's nodes that influence over the whole network
        * most important nodes
* Assortativity
    * preference of nodes to connect others that are similar
    * such as in twitter/facebook
    * happy people - tweet with other happy people
* Density - ratio of total edges vs total possible edges, how many edges a graph has
* Cliques - group of nodes that are all connected to one another
* Bridges - edge that will disconnect a graph if remove
* Clustering Coefficient - how much nodes in a graph tend to closter together

#### GANS 
* Generative Adversarial Networks (GANS) - from 2014 Ian Goodfellow
* type of unsupervised learning
* self-supervised, main use case being classifying whether something is real or not
* can be used for fraud or various SCAMS
* two submodels: generator and discriminator
* discriminatory - classification model
* real examples from training data, fake examples generated by generator 
* cryptology, drugs for making new drug chemical structures
* generates own data
* generation of data from similar distribution to fool the discriminator
* training - doesn't always converge - two players are adversaries - minimize their own loss function
* simultaneous gradient descent on both..
* minibatch - adds minibatch features 
* problems with counting - not sure how many features to include
* problems with perspective - doesn't know image from
* problems in global structure
* can have image to image translation
* multiple 'correct' answers - not just MSE - only one output
* MSE leads to averaging - generating a 'blurry' image
