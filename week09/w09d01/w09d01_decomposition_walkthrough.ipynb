#### Time Series Components
* systematic - components that are consistent and recurrent
* non-systematic - components that cannot be directly modeled 
* consist of level (average), trend(increase/decrease), seasonality (repeating cycle), noise (random variation)
* additive is addign these elements together
* multiplicative is multiplying the elements together 

Decomposition
* way of dealing with messy data 
* done with stats model

```python
# additive decomp
# can see observed, trend, seasonal (also 0), and residual (in this case is 0)
from random import randrange
from pandas import Series
from matplotlib import pyplot
from statsmodels.tsa.seasonal import seasonal_decompose
series = [i+randrange(10) for i in range(1,100)] # add random to i
result = seasonal_decompose(series, model='additive', period=1)
result.plot()
pyplot.show()
```

```python
# Multiplicative decomp
# starting with a quadratic time series
from pandas import Series
from matplotlib import pyplot
from statsmodels.tsa.seasonal import seasonal_decompose
series = [i**2.0 for i in range(1,100)]
result = seasonal_decompose(series, model='multiplicative', period=1)
result.plot()
pyplot.show()
```

```python
# decomp multiplicative on airline data
from pandas import read_csv
from matplotlib import pyplot
from statsmodels.tsa.seasonal import seasonal_decompose
df = read_csv('airline-passengers.csv', header=0, index_col=0)
result = seasonal_decompose(df, period=12, model='multiplicative') # period set to year
result.plot()
pyplot.show()
```
