Containers and Virtual Machines
* similar to virtual environments/machines
* most useful in model deployment
* Docker - tool for packing, shipping, and running applications
* goal: isolate application and dependencies to a self-contained unit to be run anywhere
* containers less secure - runs on the same kernel

Virtual Machines
* Hypervisor - software/firmware/hardware that VM's run on top of. Hypervisors run on a hostcomputer, providing the VM with RAM and CPU 
* guest machine (the VM) can run on a hosted hypervisor or bare-metal 
* hosted 
    * runs on operating system of host machine, no direct access to hardware
    * more hardware compatibility
* bare-metal
    * more performance, direct interface with underlying hardware
    * has own drivers, and interacts with components directly

Containers
* 'operating system-level virtualization'
* difference with virtual machines: containers SHARE the host's kernel with other containers
* gets its own isolated user space 
* only have their own bins and libs

Docker
* based on Linux containers
* allows creation of containers ontop of a operating system
* makes it easier to build and test portable application
* build once, and can be run anywhere - cloud, private cloud, baremetal
* Speed: fast, since just running on the kernel
* Docker hub - public images created by community for use
* modular and scalable - can link containers together to create an application
* Docker Engine
    * layer on which Docker runs
    * manages containers, images, builds 
* Docker Client
    * user interface for Docker - communicates with the Daemon (running in the background)
* Docker Daemon
    * runs on the background
    * client can run on a different machine and communicate with the daemon on the host 
* Docker Image
    * read-only templates built from a set of instructions in the Dockerfile
    * adds new layer to the image
* Union File Systems
    * what makes up a Docker Image
    * single merged directory, allows modification of certain layers while leaving original unchanged
    * duplication free, and layers are segregated
* Volumes
    * data part of a container
    * initialized when container is created
    * remains untouched even when containers are destroyed/updated/or rebuilt
    * edits need to be made directly to the volume
* Docker Containers
    * wraps the application software with everything needed to run it
    * includes operating system, application code, runtime etc.
    * built off Docker Images (read only) 
    * Container adds a writable system over the image
    * Features:
        * Namespaces - limits what the container can see and access
            * used to isolate and create a container
            * NET, PID, MNT, UTS, USER, IPC
        * Control Groups
            * ensures container only uses resources needed
            * sets what the container can use
        * Isolated Union file system
            * what allows for merging directory and makes up the image

**Removing Docker Containers,Images,Volumes,Networks**
[Reading](https://linuxize.com/post/how-to-remove-docker-images-containers-volumes-and-networks/)
* ```docker system prune``` - removes all stopped containers, dangling images, unused networks 
* -f to bypass, -a removes all unused, --volumes removes unused volumes
* containers need to be removed with --rm
* list of containers with ls -a
* remove containers by specifying name
* ```docker container rm cc3f2ff51cab cd20b396a061``` 
* removing stopped containers:
* ```docker container prune```
* for a list of stopped containers:
* ```docker container ls -a --filter status=exited --filter status=created```
* can remove based on a filter
* stop and remove all containers: 
* docker container stop $(docker container ls -aq)
* docker container rm $(docker container ls -aq)
* docker image for removing images - specify the IMAGE ID
* removing image requires removing the container first 
* can also be prune for removing dangling images
* summary:
    * done on system, container, network, image, volume
    * can: prune, --filter, ls, rm
    * if error usually means network/image/vol is being used by a container

#### Apache Spark
* big data tool for analytics, machine learning models on distributed systems, for large amount of data that won't fit in a local computer

Beginner's Guide
* for dealing with big data - coding in the cloud
* big data - volume, velocity, variety
* they come in in real time and in complex formats

Why Spark?
* unified spot to work with Big Data
* unified set of libraries
* optimizes core engine for efficiency
    * only handles storage for computation
    * not permanent storage
* libraries have wide functionalities
* fast performance, iterative, real-time analytics, graph processing, machine learning
* can work with Hadoop Distributed File System (HDFS) - especially for larger amounts of data
* Spark faster for immediate insights, processing data again and again, good for iterative computations, Machine learning, built in library, can do joins/combinations better 
* capabilities: Core (base engine), cluster management, streaming, spark SQL, GraphX, MLlib machine learning 

Good/Bad/Ugly of Spark
* Spark is like Pandas in python for cleaning and working with data, but for larger data sets that can't been run a single machine
* easy conversion from PySpark to Pandas "toPandas()"
* can easily pivot table "groupBy"
* map-side join - speeds up joins
* open source community

The Bad
* cluster management - difficult to tune and maintain
* out of memory errors
* debugging is challenging on
* Slowness of Pyspark UDFs, and requires more memory

The Ugly
* accessing array elements 
* lacks some updates, or slow to update
