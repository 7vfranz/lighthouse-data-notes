**Code guide for creating API**

**Import Flask and other packages**
```
# import Flask and jsonify
from flask import Flask, jsonify

# import Resource, Api and reqparser
from flask_restful import Resource, Api, reqparse
```
**Wrap app and api**
```
app = Flask(__name__)
api = Api(app)
```
**Code body - create a class for each feature of the app**
```
class add(Resource):
    def get(self):
        # create request parser
        parser = reqparse.RequestParser()

        # create argument 'number'
        parser.add_argument('num1', type=int, location='args')
        parser.add_argument('num2', type=int, location='args')

        # parse 'number'
        num1 = parser.parse_args().get('num1')
        num2 = parser.parse_args().get('num2')
```
**add endpoints for each class**
```
api.add_resource(add, '/add',)
```
**run the app**
```
if __name__ == '__main__':
    app.run(debug=True)
```
