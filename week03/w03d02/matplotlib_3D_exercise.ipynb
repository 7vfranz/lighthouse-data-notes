```python
# import mplot3d
from mpl_toolkits import mplot3d

# import numpy
import numpy as np

# import matplotlib
import matplotlib.pyplot as plt
```

------------

#### Task 1: Create wireframe of hyperbolic paraboloid on the domain made from intervals `<-10,10>` x `<-10,10>`

* hyperbolic paraboloid equation is: z = x*x - y*y

```python
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)

# create cartesian product
X,Y = np.meshgrid(x,y)

# set the z value
Z = (X**2) - (Y**2)
```

```python
# wireframe
fig1 = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X,Y,Z, color='b', alpha=0.2)
ax.set_title('hyperbolic paraboloid - wireframe')
```


-------------
#### Task 2: Create surface plot of eliptical paraboloid on the domain made from intervals `<-10,10>` x `<-10,10>`

* eliptical paraboloid : z = x*x + y*y

```python
# surface plot
Z2 = (X**2) + (Y**2)

fig2 = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z2, cmap='viridis')
ax.set_title('eliptical paraboloid - surface')
```
